Поиск похожего изображения с использованием Deep Features.

```python
# set product key
import graphlab
# output active product key
#graphlab.product_key.get_product_key()
```

```python
image_train = graphlab.SFrame('image_train_data/')
image_test = graphlab.SFrame('image_test_data/')
```

1. Выбрать два тестовых объекта на основе своей фамилии (написанной на латинице):

image_index1 = ASCII (decimal) 1й буквы фамилии + ASCII (decimal) 3й буквы фамилии
image_index2 = ASCII (decimal) 2й буквы фамилии + ASCII (decimal) 4й буквы фамилии

```python
image_index1 = 195   #S, p
image_index2 = 227   #z, i
```

```python
graphlab.canvas.set_target('ipynb')
#image_test[image_index1:image_index1 + 1]['image'].show()
image_test[image_index1:image_index1 + 1].show()
#image_test[image_index2:image_index2 + 1]['image'].show()
image_test[image_index2:image_index2 + 1].show()
```

2. Обучение модели собаки 

```python
dog_train=image_train[image_train['label']=='dog']
```

```python
dog_knn_model = graphlab.nearest_neighbors.create(dog_train,features=['deep_features'],
                                             label='id')
```

3. Для каждого из двух тестовых объектов из пункта 1 подсчитать ближайшего соседа, используя метод graphlab.nearest_neighbors.NearestNeighborsModel.query построенной модели.

```python
mydog1=image_test[image_index1:image_index1 + 1]
mydog2=image_test[image_index2:image_index2 + 1]
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'],'id')
mydog_neighbors1 = get_images_from_ids(dog_knn_model.query(mydog1))
mydog_neighbors2 = get_images_from_ids(dog_knn_model.query(mydog2))
```
