Кластеризация набора данных Mass-Shootings-Dataset.csv

```python
# set product key
import graphlab
#graphlab.product_key.set_product_key

# output active product key
#graphlab.product_key.get_product_key()
```

```python
data = graphlab.SFrame('Mass-Shootings-Dataset.csv')
```

```python
data.head()
```

2. Заменить в столбце Mental Health Issues значения на [0, 1, 2], где:
0 - No
1 - Yes
2 - Unclear/Unknown и другие значения

```python
data['Temp']=data['Total victims']
```

```python
data
```

```python
data['Temp'] = [1 if item == 'Yes' else 0 if item == 'No' else 2 for item in data['Mental Health Issues']]
```

```python
data['Mental Health Issues']=data['Temp']
```

```python
del data['Temp']
data
```

2. Рассчитать TF-IDF для столбца Summary.

```python
tfidf = graphlab.text_analytics.tf_idf(data['Summary'])
data['tfidf'] = tfidf
data.head()
```

3.1. На основе примера обучить модель с использованием такого набора фич:
[‘tf_idf’, ‘Total victims’, ’Mental Health Issues’] для 3 кластеров

```python
model = graphlab.kmeans.create(data,
                       features=['tfidf','Total victims','Mental Health Issues'],
                               num_clusters=3)
```

```python
model['cluster_id'].head()
```

```python
graphlab.canvas.set_target('ipynb')
model.show()
```

```python
data['cluster'] = model['cluster_id']['cluster_id']
data
```
