GENERATE & PRINT X, Y arrays

```python
import numpy as np
n = 30
X = np.random.randint(-20, 20, size=(n))
Y = np.random.randint(-20, 20, size=(n))
print ("X array example:")
print (X)
print("Y array example:")
print(Y)
```

BUBBLE SORTING X

```python
n=len(X);
m=n-1
while m>0:
    for i in range(m):
        if (X[i]>X[i+1]):
            x=X[i]
            X[i]=X[i+1]
            X[i+1]=x
        m=m-1
print ("X array sorting:")
print(X)
```
